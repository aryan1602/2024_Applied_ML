```python
import numpy as np
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import accuracy_score,confusion_matrix, ConfusionMatrixDisplay, auc, classification_report
from sklearn.model_selection import GridSearchCV
%matplotlib inline
```

```python
train = pd.read_csv('../data/train.csv')
X_train = train.drop('spam', axis=1)
y_train = train.spam
```

```python
validate_set = pd.read_csv('../data/validate.csv')
X_validate = validate_set.drop('spam', axis=1)
y_validate = validate_set.spam
```

```python
def fit_model(model, X, y):
    model.fit(X, y)
    return model
```

```python
def get_model_score(model, X, y):
    predictions = model.predict(X)
    score = accuracy_score(y, predictions)
    print("Model accuracy - "+str(score))
```

```python
def evaluate(model, X, y, predictions=None):
    if(predictions==None):
        predictions = model.predict(X)
    cm = confusion_matrix(y, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['HAM', 'SPAM'])
    disp.plot()
    plt.show()
    target_names = ["Ham", "Spam"]
    print(classification_report(y, predictions, target_names=target_names))
        

   
        
    
```

```python
def compare_models(models, X, y):
    model_scores = []
    for (key, value) in models.items():
       
        predictions = value.predict(X)
        target_names = ["Ham", "Spam"]
        clf_report = classification_report(y, predictions, target_names=target_names, output_dict=True)
        precision = clf_report["Spam"]["precision"]
        recall =clf_report["Spam"]["recall"]
        score = accuracy_score(y, predictions)
        model_scores.append((key, precision, recall, str(score)))
    
    display(pd.DataFrame(model_scores, columns = ["Model", "Precision", "Recall", "Accuracy"]))

        
    
```

## Support Vector Classifier

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
joblib.dump(svc, 'svc.joblib')
```

### Score and Evaluation on Train Set

```python
get_model_score(svc, X_train, y_train)
evaluate(svc, X_train, y_train)
```

### Score and Evaluation on Validation Set

```python
get_model_score(svc, X_validate, y_validate)
evaluate(svc, X_validate, y_validate)
```

## Gradient Boosted Trees

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)

```

### Score and Evaluation on Train Set

```python
get_model_score(clf, X_train, y_train)
evaluate(clf, X_train, y_train)
```

### Score and Evaluation on Validation Set

```python
get_model_score(clf, X_validate, y_validate)
evaluate(clf, X_validate, y_validate)
```

## Multinomial NB

```python
from sklearn.naive_bayes import MultinomialNB
nb_clf = MultinomialNB()
nb_clf.fit(X_train, y_train)
```

### Score and Evaluation on Train set

```python
get_model_score(nb_clf, X_train, y_train)
evaluate(nb_clf, X_train, y_train)
```

### Score and Evaluation on validation set

```python
get_model_score(nb_clf, X_validate, y_validate)
evaluate(nb_clf, X_validate, y_validate)
```

### Hyperparameter Tuning

```python
param_grid = {
    'alpha': [0.1, 0.01, 0.001, 0.0001], 
    'fit_prior': [True, False]  
}
gridCV = GridSearchCV(MultinomialNB(), param_grid)
gridCV.fit(X_train, y_train)
```

```python
print(gridCV.best_estimator_)
print(gridCV.best_params_)
```

```python
nb_clf = gridCV.best_estimator_
get_model_score(nb_clf, X_validate, y_validate)
evaluate(nb_clf, X_validate, y_validate)
```

## Choosing the best model

```python
test = pd.read_csv('..data/test.csv')
X_test = test.drop('spam', axis=1)
y_test = test.spam
```

```python
model_list = {
    "Multinomial Naive Bayes" : nb_clf,
    "Gradient Boosted Tree" : clf,
    "Support Vector Classifier": svc
}
compare_models(model_list, X_test, y_test)
```

##### From the above data, the best model is SVC with an accuracy of 0.9912739965095986 and precision of 0.992537
